**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Creating the function create_price_ratio that will calculate a ratio of house prices 
# in 2018 to the prices in 1998.
def create_price_ratio(df):
    year_1998 = float(df['Average_House_Price'][df['Year']==1998])
    year_2018 = float(df['Average_House_Price'][df['Year']==2018])
    ratio = [year_1998/year_2018]
    return ratio
```

```python
# Testing the function by calling it
create_price_ratio(df_group[df_group['London_Borough']=='City of London'])
```

```python
# We want to do this for all of the London Boroughs. 
# For this let's create an empty dictionary and use for loop to iterate each Borough
my_dict = {}
for borough in df_group['London_Borough'].unique():     
    bh = df_group[df_group['London_Borough'] == borough]
    my_dict[borough] = create_price_ratio(bh)
# We use the function and incorporate that into a new key of the dictionary 
print(my_dict) 
```

This dictionary contains data about the ratio of average prices for each borough between 1998 and 2018, but you can make it prettier by converting it to a DataFrame.

```python
# Making a datafrme of ratios 
df_ratios= pd.DataFrame(my_dict)
df_ratios.head()
```

```python
# transpose it and reset the index
df_ratios_T =df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
# Rename the 'index' column as 'London_Borough', and the '0' column to '2018'
df_ratios.rename(columns={'index':'London_Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```

```python
# Let's sort in descending order and select the top 10 boroughs.
top_10 = df_ratios.sort_values(by='2018',ascending=False).head(10)
print(top_10)
```

```python
# Let's plot the boroughs that have seen the greatest changes in price using bar plot
ax=top_10[['London_Borough','2018']].plot(kind='bar')
ax.set_xticklabels(top_10.London_Borough)
plt.show()
```

```python
# Another way
plt.figure(figsize=(8, 6))
x = top_10['London_Borough']
y = top_10['2018']
plt.bar(x, y, color = 'r', edgecolor = 'k')
plt.xticks(rotation=90, horizontalalignment="center")
plt.title('Top 10 London Borough with price increase', fontsize = 20)
plt.xlabel("London Borough", fontsize = 14)
plt.ylabel("Ratio", fontsize = 14)
plt.show()
```
